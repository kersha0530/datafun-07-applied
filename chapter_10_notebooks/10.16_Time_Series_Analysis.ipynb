```python
#  Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  Generate a synthetic time series dataset
date_range = pd.date_range(start="2024-01-01", periods=100, freq="D")
sales_data = np.cumsum(np.random.randn(100) * 10 + 50)  # Simulated sales trend

#  Create a DataFrame
df = pd.DataFrame({"Date": date_range, "Sales": sales_data})

#  Set Date as Index
df.set_index("Date", inplace=True)

#  Plot Time Series Data
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y="Sales", marker="o")
plt.title("Simulated Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.grid(True)
plt.show()
```
