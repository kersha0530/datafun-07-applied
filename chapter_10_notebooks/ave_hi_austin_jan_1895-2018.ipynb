```python
import pandas as pd

# Load CSV
file_path = "C:/44608 projects spring 2025/datafun-07-applied/chapter_10_examples/ave_hi_austin_jan_1895-2018.csv"
df = pd.read_csv(file_path)

# Display first few rows
print(df.head())

# Show column names and data types
print(df.info())

# Check for missing values
print(df.isnull().sum())
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")  # Convert to datetime
df.set_index("Date", inplace=True)  # Set Date as Index
print(df.head())  # Verify conversion
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Time Series Data
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y=df.columns[1], marker="o")  # Second column contains temperature/sales
plt.title("Historical Austin High Temperatures (Jan 1895-2018)")
plt.xlabel("Year")
plt.ylabel("Temperature (F)")
plt.grid(True)
plt.show()
```
